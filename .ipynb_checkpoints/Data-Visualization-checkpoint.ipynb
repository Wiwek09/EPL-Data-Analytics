<h1 style="font-style:3rem;color:orange;">EPL Football Data Analysis</h1



```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df = pd.read_csv('C:\\Users\\user\\Desktop\\New folder\\EPL_20_21.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
# Create Two New Columns
epl_df['MinutesPerMatch'] = (epl_df['Mins'] / epl_df['Matches'] ).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches'] ).astype(float)
epl_df.head()
```

```python
# Total Goals
Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

```python
# Total Penalty Goals
Total_Penalty_Goals = epl_df['Penalty_Goals'].sum()
print(Total_Penalty_Goals)
```

```python
# Total Penalty Attempts
Total_Penalty_Attempts = epl_df['Penalty_Attempted'].sum()
print(Total_Penalty_Attempts)
```

```python
# Pie-chart for penalty missed vs scored

plt.figure(figsize=(13,6))
plt_not_scored = epl_df['Penalty_Attempted'].sum() - Total_Penalty_Goals
data = [plt_not_scored,Total_Penalty_Goals]
labels = ['Penalties Missed','Penalties Scored']
color = sns.color_palette("Set2")
plt.pie(data ,labels = labels,colors = color,autopct ='%.0f')
plt.show()
```

```python
#Unique Position
epl_df['Position'].unique()
```

```python
#Forward Position Players
epl_df[epl_df['Position'] == 'FW']
```

```python
#Players from different nations
np.size(epl_df['Nationality'].unique())
```

```python
#Most Players from which Nations
nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar',figsize = (12,6),color = sns.color_palette('magma'))
```

```python
#Clubs with their maximum players in squad
epl_df['Club'].value_counts().nlargest(5).plot(kind = 'bar',color = sns.color_palette('viridis'))
```

```python
#Clubs with their minimum players in squad
epl_df['Club'].value_counts().nsmallest(5).plot(kind = 'bar',color = sns.color_palette('viridis'))
```

```python
#Players based on age group
under_20 = epl_df[(epl_df['Age'] <= 20)]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <=25)]
age25_30 = epl_df[(epl_df['Age'] >25) & (epl_df['Age'] <=30)]
above30 = epl_df[(epl_df['Age'] > 30)]
```

```python
x = np.array([under_20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above30['Name'].count()])
mylabels = ['<=20','>20 & <=25','>25 & <=30','>30']
plt.title('Players Based On Age Group',fontsize = 20)
plt.pie(x,labels = mylabels,autopct = '%.1f%%')
plt.show()
```

```python
#Total players Under 20 in each club
players_under_20 = epl_df[epl_df['Age'] <20]
plt.title('Players Under 20 Age Group',fontsize = 20)
players_under_20['Club'].value_counts().plot(kind = 'bar',color = sns.color_palette('cubehelix'))
```

```python
#Players Under 20 in ManUnited
players_under_20[players_under_20['Club'] == 'Manchester United']
```

```python
#Players Under 20 in Chelsea
players_under_20[players_under_20['Club'] == 'Chelsea']
```

```python
#Average age of Players in each club
plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club',y = 'Age',data = epl_df)
plt.xticks(rotation = 90)
```

```python
num_players = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum()) / num_players
data.sort_values(ascending = False)
```

```python
#Total assist from each club
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style = "whitegrid",color_codes = True)
ax = sns.barplot(x = 'Club',y = 'Assists',data = Assists_by_clubs.sort_values(by = 'Assists'),palette = 'Set2')
ax.set_xlabel("Club",fontsize = 30)
ax.set_ylabel("Assists",fontsize = 30)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs VS Total Assists',fontsize = 20)
```

```python
#Top 10 Assists
top_10_assists = epl_df[['Name','Club','Assists','Matches']].nlargest(n = 10,columns = 'Assists') 
top_10_assists
```

```python
#Goals by clubs
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style = "whitegrid",color_codes = True)
ax = sns.barplot(x = 'Club',y = 'Goals',data = Goals_by_clubs.sort_values(by = 'Goals'),palette = 'rocket')
ax.set_xlabel("Club",fontsize = 30)
ax.set_ylabel("Goals",fontsize = 30)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs VS Total Goals',fontsize = 20)
```

```python
#Top Goal By Players
top_10_goals = epl_df[['Name','Club','Goals','Matches']].nlargest(n = 10, columns = 'Goals')
top_10_goals
```

```python
#Top Goals Per Match
top_10_goals_per_match = epl_df[['Name','Club','GoalsPerMatch','Goals','Matches']].nlargest(n = 10, columns = 'GoalsPerMatch')
top_10_goals_per_match
```

```python
#Pie-chart - Goals with Assists and Goals without Assists
plt.figure(figsize=(14,7))
assists = epl_df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without Assists','Goals without Assists']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels , colors = color, autopct = '%.0f%%')
plt.show()
```

```python
#Top 10 players with most Yellow Cards
epl_yellow = epl_df.sort_values(by = 'Yellow_Cards',ascending = False)[:10]
plt.figure(figsize=(20,6))
plt.title("Player with the most yellow cars",fontsize = 30)
c = sns.barplot(x = epl_yellow['Name'], y = epl_yellow['Yellow_Cards'],label = 'Players',color = 'yellow')
plt.ylabel("Number of Yellow Cards")
plt.xlabel("Player")
c.set_xticklabels(c.get_xticklabels(),rotation = 45)
c
```
